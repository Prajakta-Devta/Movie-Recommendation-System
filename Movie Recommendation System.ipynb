```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
#Just for sample
convert("[{'id': 16, 'name': 'Animation'}, {'id': 35, 'name': 'Comedy'}, {'id': 10751, 'name': 'Family'}]")
convert("[{'id': 931, 'name': 'jealousy'}, {'id': 4290, 'name': 'toy'}, {'id': 5202, 'name': 'boy'}, {'id': 6054, 'name': 'friendship'}, {'id': 9713, 'name': 'friends'}, {'id': 9823, 'name': 'rivalry'}, {'id': 165503, 'name': 'boy next door'}, {'id': 170722, 'name': 'new toy'}, {'id': 187065, 'name': 'toy comes to life'}]")
```

```python
#For applying to all rows of genres column
movies['genres']=movies['genres'].apply(convert)
```

```python
#Similarly for keywords column
movies['keywords']=movies['keywords'].apply(convert)
```

```python
#Now we have to do the same task for cast column but the limitation in cast col will be we have to take only the starting 3 cast members
def convert_cast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
convert_cast(movies.iloc[0].cast)
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for s in text.split():
        y.append(ps.stem(s))
    return " ".join(y)
```

```python
#Sample after stemming
ps.stem('attires')
```

```python
#Trial for the whole paragraph after stemming
stem('when tony stark tries to jumpstart a dormant peacekeeping program, things go awry and earth’s mightiest heroes are put to the ultimate test as the fate of the planet hangs in the balance. as the villainous ultron emerges, it is up to the avengers to stop him from enacting his terrible plans, and soon uneasy alliances and unexpected action pave the way for an epic and unique global adventure. marvelcomic sequel superhero basedoncomicbook vision superheroteam duringcreditsstinger marvelcinematicuniverse 3d action adventure sciencefiction robertdowneyjr. chrishemsworth markruffalo josswhedon kevinfeige')
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer(max_features=5000,stop_words='english')
#Here stop words means that type of words which doesn't make meaning in analysis like 'in','and','on' ,etc..
#max_features means maximum no. of vectors we want to produce
```

```python
vectors = vectorizer.fit_transform(new_df['tags']).toarray()
feature_names=vectorizer.get_feature_names_out() #This line is for getting the feature names
```

```python
#Now we have got the vectors with 4809 movies & 5000 vectors so each vector is arranged accordingly
#We have to calculate the distance between each movie using Cosine Similarity
#As distance incr⬆️. Similarity decr.⬇️
```

```python
#Cosine Similarity helps to calculate the angular distance i.e 'θ' between two movies
#If the angle is 0 ° means both the movies are same
#If the angle is 5 ° means both the movies are almost similar
#If the angle is 90 ° means both the movies are not of similar type
```

```python
#We will use the sklearn library which provides in built functionality to find cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
similarity
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
#We won't consider the 0th one as it is the same movie which is provided in the argument i.e if we are provided with Avatar movie ,0th index is the movie itself
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
#Here is the final test of recommend function
recommend('The Dark Hours')
```
